K-MEAN CLUSTERING

1.All the required libraries are imported.

2.K-Mean clustering is run with 1 to 16 number of clusters and the result is plotted to determine the number of cluster.

3.Model is run again with the resulted noumber of clusters and the result is plotted.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('D:/datasets/driver-data.csv')


data.drop(['id'], axis=1, inplace = True)

from sklearn.cluster import KMeans



SSD = list() 

for K in range(1,16):
    Kmodel = KMeans(n_clusters = K, random_state=0).fit(data)
    SSD.append(Kmodel.inertia_)
    
plt.figure()
plt.plot(range(1,16), SSD)
plt.xlabel("Number of clusters") 
plt.ylabel("SSD")
plt.title("Elbow Finding Method to determine K")
plt.show()


colormap = np.array(['red', 'blue'])

Kmodel = KMeans(n_clusters = 2, random_state=0).fit(dataset) 

plt.figure(figsize = (10,8))
plt.scatter(data.mean_dist_day, data.mean_over_speed_perc, c=colormap[Kmodel.labels_], s = 40)
plt.title('K-Means Cluster')
plt.xlabel("Mean distance driven by driver per day") 
plt.ylabel("Mean percentage of time a driver was > 5 mph over the speed limit")
plt.show()
```
